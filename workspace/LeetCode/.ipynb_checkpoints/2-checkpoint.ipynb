```python
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        s=0
        lr=[]
        n=0
        lv1=0
        lv2=0
        while l1 is not None or l2 is not None:
            if l1 is None:
                lv1=0
            else:
                lv1=l1.val
                l1=l1.next
            if l2 is None:
                lv2=0
            else:
                lv2=l2.val
                l2=l2.next

            s=lv1+lv2+n
            n=0
            lr.append(s%10)
            if s>=10:
                n=1
    #         l1=l1.next
    #         l2=l2.next
        if lr[-1]==0:
            lr.append(1)
        
        sn=None
        l=len(lr)
        for i in range(l):
            sval=lr[l-i-1]
            sn=ListNode(sval,sn)
            
        return sn
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        s=0
        lr=[]
        n=0
        lv1=0
        lv2=0
        while l1 is not None or l2 is not None:
            if l1 is None:
                lv1=0
            else:
                lv1=l1.val
                l1=l1.next
            if l2 is None:
                lv2=0
            else:
                lv2=l2.val
                l2=l2.next

            s=lv1+lv2+n
            n=0
            lr.append(s%10)
            if s>=10:
                n=1
    #         l1=l1.next
    #         l2=l2.next
        if s>=10:
            lr.append(1)
        
        sn=None
        l=len(lr)
        for i in range(l):
            sval=lr[l-i-1]
            sn=ListNode(sval,sn)
            
        return sn
```
