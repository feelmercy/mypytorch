```python
#encoding=utf=8
from gensim.models import word2vec
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
import os.path
import re
import jieba
from gensim.models.word2vec import LineSentence

finput=open('F:\\study\\ml\\DataSet\\Sogou_New_Data\\news_tensite_xml.smarty\\word_segment.txt',encoding='utf-8')
foutput=open('F:\\study\\ml\\DataSet\\Sogou_New_Data\\news_tensite_xml.smarty\\corpus.txt','w',encoding='utf-8')
```

```python
  space = " "
  i = 0
```

```python
for line in finput:
    line_seg=jieba.cut(line)
    foutput.write(space.join(line_seg))
    i=i+1
    if (i % 1000 == 0):
        print("Saved " + str(i) + " articles_seg")
finput.close()
foutput.close()
```

```python
inp=open('F:\\study\\ml\\DataSet\\Sogou_New_Data\\news_tensite_xml.smarty\\corpus.txt',encoding='utf-8')
```

```python
model=Word2Vec(LineSentence(inp),size=400,window=5,min_count=5,workers=4)
```

```python
result = model.most_similar(u'部门')
```

```python
for word in result:
    print(word[0],word[1])
    
inp.close()    
```
