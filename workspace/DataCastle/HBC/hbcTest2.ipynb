```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split

from sklearn.cross_validation import *
from sklearn.grid_search import GridSearchCV

train = pd.read_csv("F:\\study\\ml\\DataSet\\Hbc_csv\\hbcTrain2.csv")
```

```python
train.head()

train['qua1']=train['qua1'].astype(int)

train.head()

column=list(train.columns)
column.remove('cMin')
column.remove('cMax')
column.remove('cAvg')

print(column)

for typeColumn in column:
    print(typeColumn)
    train[typeColumn]=train[typeColumn].astype(int)

train.head()    
```

```python
features=list(train.columns)
features.remove('orderType')
print(features)
train = train.fillna(-999)
train.head()
```

```python
train_label=train['orderType']
train_data=train.drop('orderType',axis=1)
x_train,x_val,y_train,y_val=train_test_split(train_data,train_label,test_size=0.2,random_state=100)
x_train.info()
```

```python
param = {'learning_rate' : 0.1,
        'n_estimators': 1000,
        'max_depth': 4,
        'min_child_weight': 11,
        'gamma': 0,
        'subsample': 0.7,
        'colsample_bytree': 0.75,
        'eta': 0.05,
        'silent': 0,
        'objective':'binary:logistic',
        'scale_pos_weight':1,
        'missing':[-999]}
num_round =5000
param['eval_metric'] = 'auc'

dtrain = xgb.DMatrix(x_train,y_train)
dtest = xgb.DMatrix(x_val,y_val)
evallist  = [(dtest,'eval'), (dtrain,'train')]
bst=xgb.train(param,dtrain,num_round,evallist)
```
