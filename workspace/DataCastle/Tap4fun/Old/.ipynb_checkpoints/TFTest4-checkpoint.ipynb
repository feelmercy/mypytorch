```python
text_def=[]
for col in train_feat2.columns.values:
     if 'def' in col:
            text_def.append(col)
text_def=[]
for col in train_feat2.columns.values:
     if 'def' in col:
            text_def.append(col)
text_troop=[]
for col in train_feat2.columns.values:
    if 'troop' in col :
        text_troop.append(col)
text_cavalry=[]
for col in train_feat2.columns.values:
    if 'sr_cavalry' in col :
        text_cavalry.append(col) 
text_infantry=[]
for col in train_feat2.columns.values:
    if 'sr_infantry' in col :
        text_infantry.append(col)        
```

```python
data_x=train_feat2.drop(labels=['user_id','register_time','prediction_pay_price'],axis=1)
data_y=train_feat2.prediction_pay_price
```

```python
import time
import lightgbm as lgb
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from hyperopt import hp, fmin, rand, tpe, space_eval
```

```python
X_train,X_test,y_train,y_test=train_test_split(data_x,data_y,test_size=0.3,random_state=73)
```

```python
def objective(params):
    params={
        #'max_bin':int(params['max_bin']),
        'num_leaves':int(params['num_leaves']),
        'colsample_bytree': '{:.3f}'.format(params['colsample_bytree']),
       # 'max_bin':int(params['max_bin']),
        'max_depth':int(params['max_depth']),
        'min_data_in_leaf':int(params['min_data_in_leaf']),
        #'bagging_freq':int(params['bagging_freq']),
        'lambda_l2':int(params['lambda_l2']),
        'learning_rate': '{:.3f}'.format(params['learning_rate']),
       # 'max_depth':int(params['max_depth']),
        #'min_data_in_leaf':int(params['min_data_in_leaf']),
        #'bagging_freq':int(params['bagging_freq']),
        #'lambda_l2':int(params['lambda_l2']),
        #'learning_rate':int(params['learning_rate'])
    }
    
    clf=lgb.LGBMRegressor(
        objective = 'regression', 
         metric='rmse',
        **params
    )
    
    score = cross_val_score(clf, X_train, y_train, scoring='neg_mean_squared_error',n_jobs=4,verbose=50,cv=3).mean()
    print("mse : {:.3f} <----------> params {}".format(score, params))
    return score
```

```python
space = {
    'num_leaves': hp.quniform('num_leaves', 8, 128, 4),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1.0),
    #'max_bin': hp.quniform('max_bin', 100, 128, 4),
    'max_depth': hp.quniform('max_depth', 4, 6, 1),
    'min_data_in_leaf': hp.quniform('min_data_in_leaf', 3, 6, 1),
    'lambda_l2': hp.quniform('lambda_l2', 10, 20, 2),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.1),
}

```

```python
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=500)
```
