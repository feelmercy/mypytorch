```python
import torch
import torch.nn as nn
import numpy as np
```

```python
def fizzbuzz_encode(x):
    if x % 15 == 0: return 3
    elif x % 3 == 0: return 1
    elif x % 5 == 0: return 2
    else: return 0
```

```python
def fizzbuzz_decode(x,predx):
    return [str(x),'fizz','buzz','fizzbuzz'][predx]
```

```python
def bin_encode(x,lens):
    return [x>> i & 1 for i in range(lens)][::-1]
```

```python
bin_encode(15,10)
```

```python
def bin_decode(x,lens):
    a=0
    for i in range(lens):
        if x[10-1-i] !=0:
            a+=1<<i
    return a
        
```

```python
trx=torch.Tensor([bin_encode(i,10) for i in range(101,2**10) ])
tryy=torch.LongTensor([fizzbuzz_encode(i) for i in range(101,2**10)])
```

```python
trx.size(),tryy.size()
```

```python
hidden_num=100
input_features=10
output_features=4
```

```python
model=nn.Sequential(
    nn.Linear(10,hidden_num),
    nn.ReLU(),
    nn.Linear(hidden_num,output_features)
)
```

```python
lr=0.05
loss_fn=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=lr)
```

```python
num_epochs = 10000
batch_size = 128
for epoch in range(num_epochs):
    for i in range(0, trx.shape[0], batch_size):
        batchx = trx[i:i + batch_size]
        batchy = tryy[i:i + batch_size]

        y_hat = model(batchx)
        l = loss_fn(y_hat, batchy)

        optimizer.zero_grad()
        l.backward()

        optimizer.step()
    with torch.no_grad():
        print('epoch : ', epoch, ' loss : ', loss_fn(model(trx), tryy).item())
```

```python
testx=torch.Tensor([bin_encode(i,10) for i in range(1,101)])
testx
```

```python
with torch.no_grad():
    testy=model(testx)
a=0
for i in range(1,101):
    if fizzbuzz_encode(i) ==testy.max(1)[1][i-1].item():
        a+=1
print(a/100)
    
```
