```python
dtype=torch.float
N,D_in,H,D_out=64,1000,100,10
x=torch.randn(N,D_in,dtype=dtype)
y=torch.randn(N,D_out,dtype=dtype)
w1=torch.randn(D_in,H,dtype=dtype,requires_grad=True)
w2=torch.randn(H,D_out,dtype=dtype,requires_grad=True)
learning_rate=1e-6

for t in range(500):
    y_hat=x.mm(w1).clamp(min=0).mm(w2)
    loss=(y_hat-y).pow(2).sum()
    print(t,loss.item())
    
    loss.backward()
    
    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
N, D_in, H, D_out = 64, 1000, 100, 10
x=torch.randn(N,D_in)
y=torch.randn(N,D_out)

model=torch.nn.Sequential(
    torch.nn.Linear(D_in,H),
    torch.nn.ReLU(),
    torch.nn.Linear(H,D_out),
)

loss_fn=torch.nn.MSELoss(reduction='sum')
learning_rate=1e-4
for t in range(500):
    y_hat=model(x)
    loss=loss_fn(y_hat,y)
    print(t,loss.item())
    
    model.zero_grad()
    loss.backward()
    
    
    with torch.no_grad():
        for param in model.parameters():
            param -= learning_rate * param.grad
```

```python
class TwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super().__init__()
        self.linear1=torch.nn.Linear(D_in,H)
        self.linear2=torch.nn.Linear(H,D_out)
    
    def forward(self,x):
        h_relu=self.linear1(x).clamp(min=0)
        y_pred=self.linear2(h_relu)
        return y_pred
```

```python
N, D_in, H, D_out = 64, 1000, 100, 10
x = torch.randn(N, D_in)
y = torch.randn(N, D_out)
model=TwoLayerNet(D_in, H, D_out)
y_hat=model(x)
criterion=torch.nn.MSELoss(reduction='sum')
optimizer=torch.optim.SGD(model.parameters(),lr=1e-4)
for t in range(500):
    y_hat=model(x)
    loss=criterion(y_hat,y)
    print(t,loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
def fizz_buzz_encode(i):
    if i % 15 ==0: return 3
    elif i % 5 ==0: return 2
    elif i % 3 ==0 :return 1
    else: return 0
```

```python
def fizz_buzz_decode(i,prediction):
    return [str(i),'fizz','buzz','fizzbuzz'][prediction]
```

```python
for i in range(20):
    print(fizz_buzz_decode(i, fizz_buzz_encode(i)))
```

```python
import numpy as np
```

```python
NUM_DIGITS=10
def binary_encode(d,num_digits):
    return np.array([d >>i &1 for i in range(num_digits)])[::-1]
```

```python
print(fizz_buzz_decode(1, fizz_buzz_encode(1)))
print(fizz_buzz_decode(2, fizz_buzz_encode(2)))
print(fizz_buzz_decode(5, fizz_buzz_encode(5)))
print(fizz_buzz_decode(12, fizz_buzz_encode(12)))
print(fizz_buzz_decode(15, fizz_buzz_encode(15)))
```

```python
trx=torch.Tensor([binary_encode(i,NUM_DIGITS) for i in range(101,2**NUM_DIGITS)])
tryy=torch.LongTensor([fizz_buzz_encode(i) for i in range(101,2**NUM_DIGITS)])
```

```python
trx.size()
```

```python
tryy.size()
```

```python
NUM_HIDDEN=100
model=torch.nn.Sequential(
    torch.nn.Linear(NUM_DIGITS,NUM_HIDDEN),
    torch.nn.ReLU(),
    torch.nn.Linear(NUM_HIDDEN,4)
)
```

```python
print(model)
```

```python
loss_fn=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.05)
```

```python
BATCH_SIZE=128
for epoch in range(1000):
    for start in range(0,len(trx),BATCH_SIZE):
        end=start+BATCH_SIZE
        batchx=trx[start:end]
        batchy=tryy[start:end]


        y_hat=model(batchx)
        loss=loss_fn(y_hat,batchy)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    loss=loss_fn(model(trx),tryy).item()
    print('epoch:',epoch,'loss:',loss)
```

```python
testx=torch.Tensor([binary_encode(i,NUM_DIGITS) for i in range(1,101)])
with torch.no_grad():
    testy=model(testx)

```

```python
predictions=zip(range(1,101),list(testy.max(dim=1)[1].data.tolist()))
```
