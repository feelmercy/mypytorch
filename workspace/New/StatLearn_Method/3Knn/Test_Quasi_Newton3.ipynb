```python
import numpy as np
```

```python
ffun=lambda x:100*(x[0]**2 - x[1])**2+(x[0]-1)**2
```

```python
gfun=lambda x:np.array([400*x[0]*(x[0]**2-x[1])+2*(x[0]-1),
                      -200*(x[0]**2-x[1])])
```

```python
hess=lambda x:np.array([[1200*x[0]**2-400*x[1]+2,-400*x[0]],
                        [-400*x[0],200]])
```

```python
x0=np.array([0,0])
```

```python
def dfp(ffun,gfun,hess,x0):
    maxk=1e5
    rho=0.05
    sigma=0.4
    epsilon=1e-5
    k=0
    n=np.shape(x0)[0]
    
    DK=np.linalg.inv(hess(x0))
    
    while k<maxk:
        gk=gfun(x0)
        if np.linalg.norm(gk)<epsilon:
            break
        dk=-1.0*np.dot(DK,gk)  # 海森矩阵的逆或者通过单位矩阵实现海森矩阵
        
        m=0
        mk=0
        while m<20: # ues armijo to find labmda
            if ffun(x0+rho**m*dk)<ffun(x0)+sigma*rho**m*np.dot(dk,gk):
                mk=m
                break
            m+=1
            
        ##DFP correct
        
        x=x0+rho**mk*dk
        print ("第"+str(k)+"次的迭代结果为："+str(x))
        sk=x-x0
        yk=gfun(x)-gk
        if np.dot(sk,yk)>0:
            Dy=np.dot(DK,yk)
            sy=np.dot(sk,yk)
            yDy=np.dot(np.dot(yk,DK),yk)
            DK=DK-1.0*Dy.reshape(n,1)*Dy/yDy+sk.reshape(n,1)*sk/sy
        k+=1
        x0=x
    return x0,ffun(x0),k
```

```python
dfp(ffun,gfun,hess,np.array([0,0]))
```

```python
x0,fun0,k=dfp(ffun,gfun,hess,np.array([0,0]))
print(x0,fun0,k)
```

```python
def bfgs(ffun,gfun,hess,x0):
    maxk = 1e5
    rho = 0.55
    sigma = 0.4
    gama = 0.7
    epsilon = 1e-5
    k = 0
    n=np.shape(x0)[0]
    BK=np.eye(n)  # dfp :np.linalg.inv() 求逆矩阵
    
    while k<maxk:
        gk=gfun(x0)
        if np.linalg.norm(gk)<epsilon:
            break
        dk=-1.0*np.linalg.solve(BK,gk)
        
        m=0
        mk=0
        while m<20:
            gk1=gfun(x0+rho**m*dk)
            if ffun(x0+rho**m*dk)< ffun(x0)+sigma*rho**m*np.dot(gk,dk) and np.dot(gk1.T,dk) >=gama*np.dot(gk.T,dk):
                mk=m
                break
            m+=1
        
        #BFGS correct
        x=x0+rho**mk*dk
        print("第"+str(k)+"次的迭代结果为："+str(x))
        sk=x-x0
        yk=gfun(x)-gk
        
        if np.dot(sk,yk)>0:
            Bs=np.dot(BK,sk)
            ys=np.dot(yk,sk)
            sBs=np.dot(np.dot(sk,BK),sk)
            
            BK=BK-1.0*Bs.reshape(n,1)*Bs/sBs+1.0*yk.reshape(n,1)*yk/ys
        k+=1
        x0=x
    return x0,ffun(x0),k
```

```python
bfgs(ffun,gfun,hess,np.array([3,3]))
```
