```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection  import train_test_split
import matplotlib.pyplot as plt
```

```python
# data
def create_data():
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['label'] = iris.target
    df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'label']
    data = np.array(df.iloc[:100, [0, 1, -1]])
    for i in range(len(data)):
        if data[i,-1] == 0:
            data[i,-1] = -1
    # print(data)
    return data[:,:2], data[:,-1]
```

```python
X, y = create_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
class Adaboost:
    def __init__(self,learning_rate=0.2):
        self.learning_rate=learning_rate
        
    def _em(self,axis,direct,threshold):
        if direct=='p':
            tmpl=[1 if i > threshold else -1 for i in self.x[:,axis] ]
        else:
            tmpl=[-1 if i > threshold else 1 for i in self.x[:,axis] ]
        print('tmpl: '+str(tmpl))
        tmpl2=self.y-tmpl
        tmpl3=[0 if i ==0 else 1 for i in tmpl2]
        print('tmpl3: '+str(tmpl3))
        print('D'+str(self.D))
        error=np.sum(np.multiply(tmpl3,self.D))
        print('error: '+str(error))
        return error
    
    def _predict(self,axis,threshold,direct):
        if direct=='p':
            tmpl=[1 if i > threshold else -1 for i in self.x[:,axis] ]
        else:
            tmpl=[-1 if i > threshold else 1 for i in self.x[:,axis] ]
        return tmpl
        
            
    def fit(self,x,y):
        self.x=x
        self.y=y
        self.m,self.n=x.shape
        
        self.funList=list()
        self.D=np.ones((self.m,1))/float(self.m)
        
        minError=np.inf
        minTree=None
        while minError>0.01:
            for axis in range(self.n):
                lmin=np.min(x[:,axis])
                lmax=np.max(x[:,axis])
                for threshold in np.arange(lmin,lmax,self.learning_rate):

                    e1=self._em(axis,'p',threshold)
                    e2=self._em(axis,'n',threshold)
                    print('e1: '+str(e1))
                    print('e2: '+str(e2))
                    e=e1 if e1<e2 else e2
                    d='p' if e1<e2 else 'n'
                    if minError>e:
                        minError=e
                        minTree=(axis,threshold,d,e)
            alpha=0.5*np.log((1-minError)/float(minError))
            self.funList.append((alpha,minTree))
            yi=np.array(self.y).reshape(-1,1)
            gm=np.array(self._predict(minTree[0],minTree[1],minTree[2]))
            eta=np.multiply(-alpha*yi,gm)
            Z=sum(np.multiply(self.D,np.exp(eta)))
            self.D=np.multiply(self.D,np.exp(eta))/Z

    def predict(self,x):
        sum=0
        for fun in self.funList:
            alpha=fun[0]
            tree=fun[1]
            sum+=alpha*self._predict(tree[0],tree[1],tree[2])
        return 1 if fum>0 else -1
```
