```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        results=[]
        for i in range(len(nums)):
            for j in np.arange(i+1,len(nums),1):
                if nums[i]+nums[j]==target:
                    results.append(i) 
                    results.append(j)
            if len(results)>0:
                break
        return results
```

```python
s=Solution()
```
