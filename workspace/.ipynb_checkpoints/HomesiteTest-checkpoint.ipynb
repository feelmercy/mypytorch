```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import *
from sklearn.grid_search import GridSearchCV

train=pd.read_csv("F:\\study\\ml\\DataSet\\HomesiteCompetition\\train.csv")
test=pd.read_csv("F:\\study\\ml\\DataSet\\HomesiteCompetition\\test.csv")
train.head()
train=train.drop(['QuoteNumber'],axis=1)
test=test.drop(['QuoteNumber'],axis=1)
print("################")
train.head()
```

```python
train.head()
train.loc[0,"Original_Quote_Date"]
type(train.loc[0,"Original_Quote_Date"])
```

```python
train['Date']= pd.to_datetime(train['Original_Quote_Date'])
test['Date']= pd.to_datetime(test['Original_Quote_Date'])
train['Date'].head()
```

```python
train['Year']=train['Date'].apply(lambda x:int(str(x)[:4]))
train['Month']=train['Date'].apply(lambda x:int(str(x)[5:7]))
train['weekday']=train['Date'].apply(lambda x:int(str(x)[8:10]))

test['Year']=test['Date'].apply(lambda x:int(str(x)[:4]))
test['Month']=test['Date'].apply(lambda x:int(str(x)[5:7]))
test['weekday']=test['Date'].apply(lambda x:int(str(x)[8:10]))
```

```python
train=train.drop('Date',axis=1)
test=test.drop('Date',axis=1)
train.head()
```

```python
train.isnull().any()
train.loc[:,train.isnull().any()]
train=train.fillna(-999)
test=test.fillna(-999)
train.isnull().any().any()
```

```python
np.shape(list(train['Field7'].values)+list(test['Field7'].values))
np.shape(list(train['Field7'].values))
```

```python

features = list(train.columns[2:])
a=set(train.columns)
b=set(test.columns)
a.difference(b)

for f in train.columns:
    if train[f].dtype == 'object':
        print(f)
        lbl= preprocessing.LabelEncoder()
        lbl.fit(list(train[f].values) + list(test[f].values))
        train[f]=lbl.transform(list(train[f].values))
        test[f]=lbl.transform(list(test[f].values))

train.head()        

```

```python
xgb_model = xgb.XGBClassifier()

parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['binary:logistic'],
              'learning_rate': [0.05], #so called `eta` value
              'max_depth': [6],
              'min_child_weight': [11],
              'silent': [1],
              'subsample': [0.8],
              'colsample_bytree': [0.7],
              'n_estimators': [5], #number of trees, change it to 1000 for better results
              'missing':[-999],
              'seed': [1337]}


clf = GridSearchCV(xgb_model, parameters, n_jobs=5, 
                   cv=StratifiedKFold(train['QuoteConversion_Flag'], n_folds=5, shuffle=True), 
                   scoring='roc_auc',
                   verbose=2, refit=True)
clf.fit(train[features], train['QuoteConversion_Flag'])

best_parameters, score, _ = max(clf.grid_scores_, key=lambda x: x[1])
print('Raw AUC score:', score)
for param_name in sorted(best_parameters.keys()):
    print("%s: %r" % (param_name, best_parameters[param_name]))

test_probs = clf.predict_proba(test[features])[:,1]

sample = pd.read_csv('F:\\study\\ml\\DataSet\\HomesiteCompetition\\sample_submission.csv')
sample.QuoteConversion_Flag = test_probs
sample.to_csv("F:\\study\\ml\\DataSet\\HomesiteCompetition\\xgboost_best_parameter_submission.csv", index=False)
```
