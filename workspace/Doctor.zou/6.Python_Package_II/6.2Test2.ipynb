```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def triangle_wave(size,T):
    t=np.linspace(-1,1,size,endpoint=False)
    t=abs(t)
    y=np.tile(t,T)-0.5
    x=np.linspace(0,2*np.pi*T,size*T,endpoint=False)
    return x,y
    
```

```python
x,y=triangle_wave(20,5)
x.shape
plt.plot(x,y,'g-',lw=2)
plt.show()
```
