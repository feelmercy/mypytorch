```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib as mpl
from pprint import pprint
```

```python
def restorel(sigma,u,v,K):
    m=len(u)
    n=len(v[0])
    a=np.zeros((m,n))
    for k in range(K):
        uk=u[:,k].reshape(m,1)
        vk=v[k].reshape(1,n)
        #顺序主子式累加组成A
        a+=sigma[k]*np.dot(uk,vk)
        a[a<0]=0
        a[a>255]=255
    return np.rint(a).astype('uint8')
```

```python
def restore2(sigma,u,v,K):
    m=len(u)
    n=len(v[0])
    a=np.zeros((m,n))
    for k in range(K+1):
        for i in range(m):
            a[i]+=sigma[k]*u[i][k]*v[k]
    a[a<0]=0
    a[a>255]=255
    return np.rint(a).astype('uint8')
```

```python
A=Image.open(r'F:\study\\ml\DoctorZou\06PythonPackage\6.Package\6.son.png')
output_path=r'F:\study\ml\DoctorZou\06PythonPackage\6.Package\Pic'
if not os.path.exists(output_path):
    os.mkdir(output_path)
a=np.array(A)
K=50
u_r,sigma_r,v_r=np.linalg.svd(a[:,:,0])
u_g,sigma_g,v_g=np.linalg.svd(a[:,:,1])
u_b,sigma_b,v_b=np.linalg.svd(a[:,:,2])

for k in range(1,K+1):
    print('k: ',k)
    R=restorel(sigma_r,u_r,v_r,k)
    G=restorel(sigma_g,u_g,v_g,k)
    B=restorel(sigma_b,u_b,v_b,k)
    I=np.stack((R,G,B),2)
    Image.fromarray(I).save('%s\\svd_%d.png' % (output_path,k))
    
    if k<=12:
        plt.subplot(3,4,k)
        plt.imshow(I)
        plt.axis('off')
        plt.title(u'single qua: %d' % k)
plt.suptitle(u'SVD and image',fontsize=15)
plt.tight_layout(2)
plt.subplots_adjust(top=0.9)
plt.show()
```
