```python
def read_data_to_csr(path):
    y=[]
    row=[]
    col=[]
    value=[]
    r=0
    for d in open(path):
        d=d.strip().split()
        y.append(d[0])
        d=d[1:]
        for c in d:
            k,v =c.split(':')
            col.append(int(k))
            value.append(float(v))
            row.append(r)
        r+=1
    x = scipy.sparse.csr_matrix((value,(row,col))).toarray()
    y = np.array(y)
    return x, y
            
```

```python
def show_accuracy(a, b, tip):
    acc = a.ravel() == b.ravel()
    print(acc)
    print(tip + '正确率：\t', float(acc.sum()) / a.size)
```

```python
x, y = read_data_to_csr(r'F:\study\ml\DoctorZou\11Boost\12.XGBoost\12.agaricus_train.txt')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size =0.6)
```

```python
###logistic
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_hat=lr.predict(x_test)
show_accuracy(y_hat,y_test,'LR')
```

```python
###xgb
data_train=xgb.DMatrix(x_train,label=y_train)
data_test=xgb.DMatrix(x_test,label=y_test)
watch_list=[(data_test,'eval'),(data_train,'train')]
param={'max_depth':3,'eta':0.01,'silent':0,'objective':'binary:logistic'}
```

```python
bst=xgb.train(param,data_train,num_boost_round=4,evals=watch_list)
y_hat=bst.predict(data_test)
```
