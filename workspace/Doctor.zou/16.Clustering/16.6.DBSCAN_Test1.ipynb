```python
X1, y1 = datasets.make_circles(n_samples=2000, factor=.6, noise=.02)
X2, y2 = datasets.make_blobs(n_samples=400, n_features=2, centers=[[1.2, 1.2]], cluster_std=[[.1]], random_state=9)
X = np.concatenate((X1, X2))
eps = 1
MinPts = 10
```

```python
eps = 0.2
MinPts = 5

ptses = []
dist = euclidean_distances(X)
```

```python
for row in dist:
    density=np.sum(row<eps)
    pts=0
    if density>MinPts:
        pts=1
    elif density >1:
        pts=2
    else:
        pts=0
    ptses.append(pts)
```

```python
cp_index=[]
for i ,v in enumerate(ptses):
    if v!=0:
        cp_index.append(i)
```

```python
corePoints=X[cp_index,:]
coreDist=euclidean_distances(corePoints)
```
