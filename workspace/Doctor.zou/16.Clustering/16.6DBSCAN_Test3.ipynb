```python
from sklearn import datasets
import numpy as np
import random
import matplotlib.pyplot as plt
import time
import copy
from sklearn.metrics import euclidean_distances
%matplotlib auto
```

```python
x1,y1=datasets.make_circles(n_samples=2000,noise=0.02,factor=0.6,random_state=5)
x2,y2=datasets.make_blobs(n_features=2,n_samples=500,centers=[[1.2,1.2]]
                          ,cluster_std=0.1,random_state=5)
data=np.vstack((x1,x2))
```

```python
data.shape
```

```python
eps=0.2
minpts=10
```

```python
dist=euclidean_distances(data)
ptype_list=[]
for row in dist:
    ptype=0
    density=np.sum(np.where(row<eps))
    if density >minpts:
        ptype=1
    elif density >1 :
        ptype=2
    else :
        ptype=0
    ptype_list.append(ptype)
```

```python
cpt_index=[]
for k,v in enumerate(ptype_list):
    if v >0:
        cpt_index.append(k)
```

```python
cpt_data=data[cpt_index,:]
cpt_dist=euclidean_distances(cpt_data)
```
