```python
def linreg_net(x,w,b):
    return torch.mm(x,w)+b
```

```python
def squared_loss(y_hat,y):
    return (y_hat.reshape(-1,1)-y)**2
```

```python
def sgd(params,lr,batch_size):
    for p in params:
        p.data -=lr*p.grad/batch_size
```

```python
w=torch.Tensor(np.random.normal(0,0.01,(x.shape[1],1)))
b=torch.Tensor([1.])
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

```python
features=torch.Tensor(x)
labels=torch.Tensor(y)
dataset=torch.utils.data.TensorDataset(features,labels)
```

```python
epoch_num=10000000
lr=0.0003
batch_size=512
```

```python
w=torch.Tensor(np.random.normal(0,0.01,(x.shape[1],1)))
b=torch.Tensor([1.])
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)

for e in range(epoch_num):
    for x,y in torch.utils.data.DataLoader(dataset,512,shuffle=True):
        l=squared_loss(linreg_net(x,w,b),y).mean()
#         print(linreg_net(x,w,b))
#         print('l:',l.item())
        if w.grad is not None:
#             print('---------------')
#             print(w.grad.data)
#             print('---------------')
            for p in [w,b]:
                p.grad.data.zero_()

        l.backward()
        sgd([w,b],lr,512)
    if (e+1) %1000 ==0:
        print('loss :' ,l.item())
    
```
