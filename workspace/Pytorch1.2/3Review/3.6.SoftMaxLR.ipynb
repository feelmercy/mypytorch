```python
import torch
import torchvision
import numpy as np
import torch.utils.data as Data
from torchvision import transforms
```

```python
batch_size=256
mnist_train=torchvision.datasets.FashionMNIST(
    root='F:\study\ml\DataSet\FashionMNIST',train=True,
    download=True,transform=transforms.ToTensor())
mnist_test=torchvision.datasets.FashionMNIST(
    root='F:\study\ml\DataSet\FashionMNIST',train=False,
    download=True,transform=transforms.ToTensor())
```

```python
train_iter=Data.DataLoader(mnist_train,batch_size,shuffle=True)
test_iter=Data.DataLoader(mnist_test,batch_size,shuffle=False)
```

```python
def softmax(x):
    x_exp=x.exp()
    x_exp_sum=x_exp.sum(dim=1,keepdim=True)
    return x_exp/x_exp_sum
    
```

```python
a=torch.Tensor([[1,2,3],[4,5,6]])
softmax(a)
```

```python
def snet(x,w,b):
    return softmax(torch.mm(x.view(-1,784),w)+b)
```

```python
def cross_entropy(y_hat,y):
    return-torch.log(y_hat.gather(1,y.view(-1,1)))
```

```python
b=torch.Tensor([[0.1,0.2,0.7],[0.3,0.4,0.3]])
c=torch.tensor([1,2],dtype=torch.long)
print(softmax(b))
print(-torch.log(softmax(b)))
print(-torch.log(b))
cross_entropy(b,c)
```

```python
def sgd(params,lr,batch_size):
    for p in params:
        p.data -= lr*p.grad/batch_size
```

```python
def evaluate_acc(data_iter):
    acc,n=0,0
    for x,y in data_iter:
        acc+=(snet(x,w,b).argmax(dim=1)==y).sum().item()
        n+=len(y)
    return acc/n
```

```python
w=torch.Tensor(np.random.normal(0,0.01,(784,10)))
b=torch.zeros((1,10),dtype=torch.float32)
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

```python
lr=0.03
params=[w,b]
loss=cross_entropy
for i in range(5):
    train_l,train_acc,test_acc,n=0,0,0,0
    for x,y in train_iter:
        y_hat=snet(x,w,b)
        l=loss(y_hat,y).sum()
#         l=loss(y_hat,y)
        print(l)
        
        if params[0].grad is not None:
            for p in params:
                p.grad.data.zero_()
        l.backward()
        sgd(params,lr,batch_size)
        train_l+=l
        train_acc+=(snet(x,w,b).argmax(dim=1)==y).sum().item()
        n+=len(y)
    test_acc=evaluate_acc(test_iter)
    print('eposh %d,loss %.3f,train acc %.3f,test acc %.3f' % (
    i+1,train_l/n,train_acc/n,test_acc))
```
